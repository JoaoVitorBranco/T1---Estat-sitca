## 0.3 - Definindo funções

```python
def plotaHist(data,x,titulo,xlabel,ylabel, **kwargs):
    ax = sns.histplot(data=data,x=x,shrink=0.9,multiple='stack', **kwargs)
    ax.set_title(titulo, fontsize=20)
    ax.set_xlabel(xlabel, fontsize=20)
    ax.set_ylabel(ylabel, fontsize=15)
    ax.tick_params(labelsize=15)
    ax.axvline(data[x].mean(), ls='-.',color='red',alpha=0.7)
    ax.annotate(f'Media = {round(data[x].mean(),1)}', xy=(data[x].mean()*1.1,data[x].value_counts().max()*0.9), color='red', fontsize=15,alpha=0.7)
    ax.grid(False)
    return ax

def plotaBox(data,x,titulo,xlabel,ylabel, mostra_media = True, **kwargs):
    ax = sns.boxplot(data=data,x=x, **kwargs)
    ax.set_title(titulo, fontsize=20)
    ax.set_xlabel(xlabel, fontsize=20)
    ax.set_ylabel(ylabel, fontsize=15)
    ax.tick_params(labelsize=15)
    ax.grid(False)
    plt.xticks(rotation=0)
    if(mostra_media):
        ax.axvline(data[x].mean(),ls='-.',color='red',alpha=0.7)
        ax.annotate(f'Media = {round(data[x].mean(),1)}',xy=(data[x].mean()*1.1,0.45),fontsize=15,color='red',alpha=0.7)
        ax.annotate(f'Mediana = {round(data[x].median(),1)}',xy=(data[x].median()*1.1,-0.45),fontsize=15,c='k',alpha=0.7)
    return ax

def plotaScatt(data,x,y,titulo,xlabel,ylabel):
    ax = sns.regplot(data=data,x=x,y=y,ci=0)
    ax.set_title(titulo, fontsize=20)
    ax.set_xlabel(xlabel, fontsize=20)
    ax.set_ylabel(ylabel, fontsize=15)
    ax.tick_params(labelsize=15)
    ax.grid(False)
    return ax

def plotaBarra(data, x, y, titulo, xlabel, ylabel):
    ax = sns.barplot(data=data, x=x, y=y)
    ax.set_title(titulo, fontsize=20)
    ax.set_xlabel(xlabel, fontsize=20)
    ax.set_ylabel(ylabel, fontsize=15)
    ax.tick_params(labelsize=15)
    ax.grid(False)
    return ax


def tableOfFrequency(series,tradutor=False):
    dados = pd.DataFrame(series.value_counts())
    dados.sort_index(inplace=True)

    if tradutor:
        dados[''] = dados.index.to_series().map(tradutor)
        dados.set_index('',inplace=True)
        
    
    total = dados['count'].sum()
    dados.loc['total'] = total
    dados['%'] = round((dados['count']/total)*100,2)
    
    return dados

def relQualQuant(data, cQuant, cQual, titulo, xlabel, ylabel):
    plotaBox(data=data, x=cQual, titulo=titulo, xlabel=xlabel, ylabel=ylabel, y = cQuant, mostra_media=False)
    display(data.groupby(cQual)[cQuant].describe())

def relQual2(data, c1, c2, titulo):
    print(titulo)
    dados = pd.crosstab(data[c1], data[c2], normalize='index').apply(lambda x: round(x * 100,2)).astype(str) + ' %'
    display(dados)
```
